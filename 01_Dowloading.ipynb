### üåä Downloading Copernicus Marine Data 

Sources :
- üåê [*Baltic Sea Physics Reanalysis*](https://data.marine.copernicus.eu/product/BALTICSEA_MULTIYEAR_PHY_003_011/description)
- üß™ [*Baltic Sea Biogeochemistry Reanalysis*](https://data.marine.copernicus.eu/product/BALTICSEA_MULTIYEAR_BGC_003_012/description)

There are two ways you can download data from the Copernicus Marine Data Store:

#### 1. üíª Manual Download 

For the [*Baltic Sea Physics Reanalysis*](https://data.marine.copernicus.eu/product/BALTICSEA_MULTIYEAR_PHY_003_011/description) and [*Baltic Sea Biogeochemistry Reanalysis*](https://data.marine.copernicus.eu/product/BALTICSEA_MULTIYEAR_BGC_003_012/description) :

0. Login or create an account
1. Click on ‚ÄúData access‚Äù and choose the desired time resolution (I selected "Yearly").
2. Download the datasets using the ‚ÄúFrom‚Äù or ‚ÄúBrowse‚Äù buttons. 

#### 2. üì° Using the Python API 

For the [*Baltic Sea Wave Hindcast*](https://data.marine.copernicus.eu/product/BALTICSEA_MULTIYEAR_WAV_003_015/description), I used the Python API to download it programmatically, which is an efficient method for automating the process and handling large datasets. Since this dataset is available only in hourly resolution, I had to calculate the yearly mean to match the annual resolution of the other datasets. See the *Calculating_annual_mean* Notebook.
