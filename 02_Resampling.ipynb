```python
import xarray as xr
import rasterio
from rasterio.transform import from_origin
import os
```

```python
# Define the directory where you want to save the GeoTIFF
output_directory = "C:/Users\PC\Documents\Ecole/2A\Stage_2A\RISE\Example\Angiosperms-model"  # Replace with your desired directory path
output_file = os.path.join(output_directory, "chl_2019.tif")  # GeoTIFF output file path

# Check if the directory exists, if not, create it
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
```

```python
ds = xr.open_dataset('filtered_BAL-MYP-ERGOM_BGC-AnnualMeans-2019.nc')
ds
```

```python
# Extract the 2D array
chl = ds['chl'].values  
```

```python
chl[0,22,150,50]
```

```python
lon = ds['lon'].values
lat = ds['lat'].values
```

```python
# Define the transform (adjust based on your dataset)
transform = from_origin(lon.min(), lat.max(), (lon[1] - lon[0]), (lat[0] - lat[1]))  # Adjust according to your data

# Define metadata
metadata = {
    'driver': 'GTiff',
    'count': 1,  # Single band
    'dtype': 'float32',
    'crs': 'EPSG:4326',  # Define CRS (update as needed)
    'transform': transform,
    'width': chl.shape[1],
    'height': chl.shape[0],
    'nodata': -9999  # Example NoData value
}

# Write the data to a GeoTIFF
with rasterio.open('chl_2019.tif', 'w', **metadata) as dst:
    dst.write(chl, 1)  # Write data to the first band
```
